```python
#   Automatic Differentiation
import tensorflow as tf
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

#   To prevent a total meltdown of my GPU, restrictions needed to be made.
rtx2080 = tf.config.experimental.list_physical_devices('GPU')
for num in rtx2080:
    tf.config.experimental.set_memory_growth(num, True)

x = tf.range(4, dtype=tf.float32)
x
```

```python
class Mona:
    def __init__(self, mu, sd):
        self.sd = sd
        self.mu = mu
    
    def normdist(tiger):
        x = np.arange(-2,2,0.01)
        y = norm.pdf(x, tiger.mu, tiger.sd)      
        return y


res = Mona(0,1).normdist()
x = np.arange(-2,2,0.01)
plt.plot(x, res)
plt.show()
```
